Exercise 1: Calculate the multiplication and sum of two numbers 

Given two integer numbers return their product only if the product is greater than 1000, else return their sum.



image.png

```python
# YOUR CODE HERE
def multiplication_or_sum(num1, num2):
    product = num1 * num2
    return product if product > 1000 else num1 + num2

print('The result is', multiplication_or_sum(20, 30))
print('The result is', multiplication_or_sum(40, 30))
```

Exercise 2: Print the sum of the current number and the previous number

Write a program to iterate the first 10 numbers and in each iteration, print the sum of the current and previous number.

image.png

```python
# YOUR CODE HERE
def print_num_prev_and_sum():
    prev_num = 0
    for i in range(10):
        curr_sum = prev_num + i
        print('Current Number', i, 'Previous Number', prev_num, 'Sum:', curr_sum)
        prev_num = i

print_num_prev_and_sum()
```
